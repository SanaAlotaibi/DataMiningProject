### 3-Description and Analysis of the Results of Pre-processing Techniques

In this section, we analyze the impact of the pre-processing steps applied to our dataset to ensure it is clean and ready for further analysis.

#### **1. Handling Missing Values**  
After checking for missing values, we found that **all columns have complete data**, meaning there were no missing values that needed imputation or removal. This helped us proceed without needing to fill in or remove any records.

#### **2. Outlier Detection and Removal**  
To check for outliers, we used **summary statistics and boxplots** on numerical columns like **stress level, working hours per week, commute time, and salary**.  
- **Stress Level (0-10)**: The values are within a reasonable range, so no outliers were removed.  
- **Working Hours (35-90 hours)**: Although some employees have long working hours, these values are realistic for certain job roles, so we decided not to remove them.  
- **Commute Time (0-3 hours)** and **Sleep Hours (4-9 hours)**: The values are within an expected range.  

Since no extreme or unrealistic outliers were found, **no data points were removed**.  

#### **3. Feature Scaling (Normalization/Standardization)**  
For now, we have kept the original values for numerical features. However, for machine learning models, we might need to normalize variables like **salary and working hours** to avoid large numerical differences affecting the model.

#### **4. Encoding Categorical Variables**  
Our dataset contains several categorical columns, such as **Gender, Marital Status, Job Role, Company Size, and Location**.  
- To prepare these for analysis, we converted them into numerical values using **One-Hot Encoding or Label Encoding**.  
- For example, **Gender ("Male", "Female", "Non-Binary")** was encoded into numerical categories.

#### **5. Feature Selection**  
Since our dataset contains **30 columns**, we checked for highly correlated variables that might not add much value to the analysis.  
- The **ID column** was removed since it does not contribute to predictions.  
- Features like **Job Satisfaction, Work-Life Balance, and Stress Level** were kept as they are important for our study.
